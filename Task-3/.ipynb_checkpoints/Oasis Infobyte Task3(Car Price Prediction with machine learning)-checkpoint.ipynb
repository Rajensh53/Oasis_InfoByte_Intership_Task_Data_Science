### Replace Columns Value into Numarics

```python
#replace value of fueltype column 
car_data.replace({'fueltype':
                  {'gas': 0,
                  'diesel':1}}, inplace =True)

#replace value of aspiration column 
car_data.replace({'aspiration':
                  {'std': 0,
                  'turbo':1}}, inplace =True)

#replace value of doornumber column 
car_data.replace({'doornumber':
                  {'four': 4,
                  'two':2}}, inplace =True)

#replace value of doornumber column 
car_data.replace({'enginelocation':
                  {'front': 0,
                  'rear':1}}, inplace =True)

#replace value of drivewheel column 
car_data.replace({'drivewheel':
                  {'fwd': 0,
                  'rwd':1,
                  '4wd':2}}, inplace =True)

#replace value of cylindernumber column 
car_data.replace({'cylindernumber':
                  {'four': 4,
                  'six':6,
                  'five':5,
                  'three': 3,
                  'two': 2,
                  'eight': 8,
                  'twelve': 12}}, inplace =True)

#replace value of fuelsystem column 
car_data.replace({'fuelsystem':
                  {'mpfi': 0,
                  '2bbl':1,
                  'idi':2,
                  '1bbl': 3,
                  'spdi': 4,
                  '4bbl': 5,
                  'mfi': 6,
                  'spfi': 7}}, inplace =True)

#replace value of enginetype column 
car_data.replace({'enginetype':
                  {'ohc': 0,
                  'ohcf':1,
                  'ohcv':2,
                  'dohc': 3,
                  'l': 4,
                  'rotor': 5,
                  'dohcv': 6}}, inplace =True)

#replace value of carbody column 
car_data.replace({'carbody':
                  {'sedan': 0,
                  'hatchback':1,
                  'wagon':2,
                  'hardtop': 3,
                  'convertible': 4}}, inplace =True)
```

## Seperate the data into Train and Test data

```python
X = car_data.drop(['CarName', 'car_ID', 'price'],axis=1)
y = car_data.price
```

```python
print(X.head())
print(y.head())
```

### Now spliting Train and Test data

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size=0.1, random_state=2)
```

```python
X.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
# Mode Evaluation
#Prediction on training data
train_data_prediction = model.predict(X_train)
```

```python
#R square Error
error_score = metrics.r2_score(y_train, train_data_prediction)
print("R score Error", error_score)
```

```python
# Visualize the actual prices and predictes price
plt.scatter(y_train, train_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices vs Predicted Prices")
plt.show
```

```python
#Prediction on testing data
test_data_prediction = model.predict(X_test)
```

```python
#R square Error
error_score = metrics.r2_score(y_test, test_data_prediction)
print("R score Error", error_score)
```

```python
# Visualize the actual prices and predictes price
plt.scatter(y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices vs Predicted Prices")
plt.show
```
